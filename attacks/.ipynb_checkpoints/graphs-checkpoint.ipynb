```python
import matplotlib.pyplot as plt
```

```python
## S2

# t values
t = [0,1,2,3]
# corresponding fraction of insert items
frac = [50,50,0,0]
#expected fraction according to the math
expected = [50,50,0,0]
# increase in cardinality
card = [113,25,0,0]
 
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(t, frac, 'g-')
ax1.plot(t, expected, 'g.')
ax2.plot(t, card, 'b-')

ax1.set_xlabel('t')
ax1.set_ylabel('Fraction of inserted items', color='g')
ax2.set_ylabel('Average cardinality increase', color='b')

plt.show()

fig.savefig("../docs/HLL_exp_bad/figures/graph_S2.pdf", bbox_inches='tight')
```

```python
## S4

# t values
t = [0,1,2,3]
# corresponding fraction of insert items
frac = [50,50,0,0]
#expected fraction according to the math
expected = [50,50,0,0]
# increase in cardinality
card = [0,0,0,0]
 
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(t, frac, 'g-', label = 'actual')
ax1.plot(t, expected, 'g..', label = 'expected')
ax1.legend()
ax2.plot(t, card, 'b-')

ax1.set_xlabel('t')
ax1.set_ylabel('Fraction of inserted items', color='g')
ax2.set_ylabel('Average cardinality increase', color='b')

plt.show()

fig.savefig("../docs/HLL_exp_bad/figures/graph_S4.pdf", bbox_inches='tight')
```
